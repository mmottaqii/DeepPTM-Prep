```python
import pandas as pd
import re

# Define the chemical concentration (list)
# Define the PTM representation explicitly
ptm_representation = '(ph)'  # Modify this to '(ac)' or any other representation as needed
ptm_type = 'phosphorylation'  # Specify the PTM type for the new column
# Load the dataset
df = pd.read_csv('curves_subset_10_kinase_2KI.txt', sep='\t')

# Define the list of concentrations
concentrations = [0, 0.03, 0.3, 1, 3, 10, 30, 100, 300, 1000, 10000]




# Define a function to extract experiment details
def extract_experiment_details(experiment_str):
    parts = experiment_str.split('_')
    experiment_type = parts[0]
    cell_line = parts[1]
    chemical_name = parts[2]
    time_unit = parts[3][-1]  # Get the last character which indicates the time unit (m or h)
    timepoint = int(re.findall(r'\d+', parts[3])[0])
    
    # Convert timepoint to minutes if it is in hours
    if time_unit == 'h':
        timepoint *= 60
        
    return experiment_type, cell_line, chemical_name, timepoint



# Define a function to split modified sequence and get PTM site information
def get_ptm_sites(modified_sequence, raw_sequence, ptm_representation='(ph)'):
    # Remove '(ac)' from the start
    modified_sequence = modified_sequence.replace('(ac)', '', 1)  

    ptm_sites = []
    index_offset = 0  # Track how many characters we've removed

    for i, char in enumerate(modified_sequence):
        if char == '(':  # Start of a potential PTM site
            if modified_sequence[i:i+len(ptm_representation)] == ptm_representation:
                ptm_residue = raw_sequence[i - index_offset - 1]  # Get residue before (ph)
                ptm_index = i - index_offset  # Calculate index in raw sequence
                ptm_sites.append((ptm_index, ptm_residue))  # Store as one-based index

                index_offset += len(ptm_representation)  # Update the offset

    return ptm_sites





# Prepare a list to hold the reformatted data
reformatted_data = []

# Process each row in the dataset
for _, row in df.iterrows():
    experiment_type, cell_line, chemical_name, timepoint = extract_experiment_details(row['Experiment'])
    ptm_sites = get_ptm_sites(row['Modified sequence'], row['Sequence'], ptm_representation)
    
    for idx, concentration in enumerate(concentrations, start=1):
        for ptm_site in ptm_sites:
            reformatted_data.append({
                'Sequence': row['Sequence'],
                'PTM Index': ptm_site[0],  # One-based index
                'PTM Residue': ptm_site[1],
                'PTM type': ptm_type,  # Add the PTM type column
                'Sequence length': row['Length'],
                
                'Proteins': row['Proteins'],
                'Leading proteins': row['Leading proteins'],
                'Protein names': row['Protein names'],
                'Gene names': row['Gene names'],
                'Missed cleavages': row['Missed cleavages'],
                
                'Experiment': row['Experiment'],
                'Experiment Type': experiment_type,
                'Cell Line': cell_line,
                'Timepoint (min)': timepoint,
                'Chemical Name': chemical_name,
                'Dosage (µg)': concentration,
                'N duplicates': row['N duplicates'],
                
                'Signal Intensity': row[f'TMT Channel Normal {idx}'],
                'Signal Ratio': row[f'TMT Channel Ratio {idx}'],
                'Curve signal': row['Curve signal'],
                'Max Score': row['Max Score'],
                'Min PEP': row['Min PEP'],
                
                'Log EC50': row['Log EC50'],
                'Curve slope': row['Curve slope'],
                'Curve top': row['Curve top'],
                'Curve bottom': row['Curve bottom'],
                'R2': row['R2'],
                'Curve RMSE': row['Curve RMSE'],
                'Log EC50 error': row['Log EC50 error'],
                'Curve slope error': row['Curve slope error'],
                'Curve top error': row['Curve top error'],
                'Curve bottom error': row['Curve bottom error'],
                'EC50': row['EC50'],
                'pEC50': row['pEC50'],
                'Curve effect size': row['Curve effect size']
            })

# Create a new DataFrame from the reformatted data
reformatted_df = pd.DataFrame(reformatted_data)

# Save the reformatted data to a new CSV file
#reformatted_df.to_csv('reformatted_ptm_7.csv', index=False)

print("Reformatted data has been saved to 'reformatted_ptm_data.csv'")
```

```python
import pandas as pd
import pickle

# Load the TSV file
file_path = '/home/vmottaqi/decryptm/curves/processed_curves/modified_processed_curves/gene_uniprotID_mapping.tsv'  # Replace with the path to your TSV file
df = pd.read_csv(file_path, sep='\t')

# Initialize an empty dictionary to store the mapping
protein_mapping = {}

# Process each row in the DataFrame
for _, row in df.iterrows():
    # Extract the necessary columns
    protein_id = row['Entry']
    protein_name = row['Protein names']
    gene_names = row['Gene Names']
    
    # Combine the gene names into a list
    genes = set(gene_names.split())
    
    # Check if the protein is already in the dictionary
    if protein_id in protein_mapping:
        # Update the gene names list
        existing_genes = set(protein_mapping[protein_id][0])
        updated_genes = existing_genes.union(genes)
        protein_mapping[protein_id][0] = list(updated_genes)
    else:
        # Add a new entry to the dictionary
        protein_mapping[protein_id] = [list(genes), protein_name]

# Save the dictionary as a pickle file
output_pickle_file = '/home/vmottaqi/decryptm/curves/processed_curves/modified_processed_curves/gene_uniprotID_mapping.pkl'  # Replace with the desired output path
with open(output_pickle_file, 'wb') as f:
    pickle.dump(protein_mapping, f)

print(f"Protein mapping dictionary has been saved to '{output_pickle_file}'")
```

```python
import os
import pandas as pd
import pickle

# Load the mapping dictionary
mapping_pickle_file = '/home/vmottaqi/decryptm/curves/processed_curves/gene_uniprotID_mapping.pkl'  # Replace with the path to your pickle file
with open(mapping_pickle_file, 'rb') as f:
    protein_mapping = pickle.load(f)

# Directory containing the .pkl files
directory = '/home/vmottaqi/decryptm/curves/processed_curves/modified_processed_curves'  # Replace with your directory path

# Initialize a list to store error details
errors = []

# Process each .pkl file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.pkl'):
        file_path = os.path.join(directory, filename)
        try:
            # Load the .pkl file into a DataFrame
            df = pd.read_pickle(file_path)
            
            # Ensure the columns exist
            if 'Gene names' not in df.columns:
                df['Gene names'] = None
            if 'Leading proteins' not in df.columns:
                df['Leading proteins'] = None
            if 'Protein names' not in df.columns:
                df['Protein names'] = None
            
            # Process each row in the DataFrame
            for index, row in df.iterrows():
                gene_names = row['Gene names']
                leading_protein = row['Leading proteins']
                protein_ids = row['Protein IDs'] if 'Protein IDs' in row else None
                gene_name = None
                protein_id = None
                protein_name = None

                if pd.isna(gene_names) or gene_names == '':
                    # Check leading protein or protein IDs
                    if pd.isna(leading_protein) or leading_protein == '':
                        if pd.isna(protein_ids) or protein_ids == '':
                            errors.append(f"File '{filename}' row {index} has no valid gene names, leading proteins, or protein IDs")
                            continue  # Skip to the next row
                        else:
                            # Check each protein ID
                            found = False
                            for pid in protein_ids.split(';'):
                                pid = pid.strip()
                                if pid in protein_mapping:
                                    gene_list = protein_mapping[pid][0]
                                    gene_name = gene_list[0] if gene_list else None
                                    protein_id = pid
                                    protein_name = protein_mapping[pid][1]
                                    df.at[index, 'Gene names'] = gene_name
                                    df.at[index, 'Leading proteins'] = protein_id
                                    df.at[index, 'Protein names'] = protein_name
                                    found = True
                                    break
                            if not found:
                                # Set the first protein ID as the leading protein
                                first_pid = protein_ids.split(';')[0].strip()
                                df.at[index, 'Leading proteins'] = first_pid
                                errors.append(f"File '{filename}' row {index} protein IDs not found in dictionary, set first protein ID '{first_pid}' as leading protein")
                                continue  # Skip to the next row
                    else:
                        if leading_protein not in protein_mapping:
                            errors.append(f"File '{filename}' row {index} leading protein '{leading_protein}' not found in dictionary")
                            continue  # Skip to the next row
                        gene_list = protein_mapping[leading_protein][0]
                        gene_name = gene_list[0] if gene_list else None
                        protein_name = protein_mapping[leading_protein][1]
                        df.at[index, 'Gene names'] = gene_name
                        df.at[index, 'Protein names'] = protein_name
                else:
                    # Use the first gene name in case of multiple gene names
                    gene_name = gene_names.split(';')[0].strip()
                    found = False
                    for pid, (gene_list, protein) in protein_mapping.items():
                        if gene_name in gene_list:
                            protein_id = pid
                            protein_name = protein
                            df.at[index, 'Leading proteins'] = protein_id
                            df.at[index, 'Protein names'] = protein_name
                            found = True
                            break
                    if not found:
                        errors.append(f"File '{filename}' row {index} gene name '{gene_name}' not found in dictionary")
                        continue  # Skip to the next row
            
            # Save the updated DataFrame back to the .pkl file
            df.to_pickle(file_path)
        
        except Exception as e:
            errors.append(f"Error processing file '{filename}': {str(e)}")

# Print all errors
if errors:
    print("Errors encountered during processing:")
    for error in errors:
        print(error)
else:
    print("All files processed successfully.")
```

```python
import os
import pandas as pd
import pickle
from joblib import Parallel, delayed

# Load the mapping dictionary
mapping_pickle_file = '/home/vmottaqi/decryptm/curves/processed_curves/gene_uniprotID_mapping.pkl'  # Replace with the path to your pickle file
with open(mapping_pickle_file, 'rb') as f:
    protein_mapping = pickle.load(f)

# Directory containing the .pkl files
directory = '/home/vmottaqi/decryptm/curves/processed_curves/modified_processed_curves/test'  # Replace with your directory path

# Initialize a list to store error details
errors = []

def process_file(filename):
    file_path = os.path.join(directory, filename)
    local_errors = []
    try:
        # Load the .pkl file into a DataFrame
        df = pd.read_pickle(file_path)

        # Ensure the columns exist
        if 'Gene names' not in df.columns:
            df['Gene names'] = None
        if 'Leading proteins' not in df.columns:
            df['Leading proteins'] = None
        if 'Protein names' not in df.columns:
            df['Protein names'] = None

        # Process each row in the DataFrame
        for index, row in df.iterrows():
            gene_names = row['Gene names']
            leading_protein = row['Leading proteins']
            protein_ids = row['Protein IDs'] if 'Protein IDs' in row else None
            gene_name = None
            protein_id = None
            protein_name = None

            if pd.isna(gene_names) or gene_names == '':
                # Check leading protein or protein IDs
                if pd.isna(leading_protein) or leading_protein == '':
                    if pd.isna(protein_ids) or protein_ids == '':
                        local_errors.append(f"File '{filename}' row {index} has no valid gene names, leading proteins, or protein IDs")
                        continue  # Skip to the next row
                    else:
                        # Check each protein ID
                        found = False
                        for pid in protein_ids.split(';'):
                            pid = pid.strip()
                            if pid in protein_mapping:
                                gene_list = protein_mapping[pid][0]
                                gene_name = gene_list[0] if gene_list else None
                                protein_id = pid
                                protein_name = protein_mapping[pid][1]
                                df.at[index, 'Gene names'] = gene_name
                                df.at[index, 'Leading proteins'] = protein_id
                                df.at[index, 'Protein names'] = protein_name
                                found = True
                                break
                        if not found:
                            # Set the first protein ID as the leading protein
                            first_pid = protein_ids.split(';')[0].strip()
                            df.at[index, 'Leading proteins'] = first_pid
                            local_errors.append(f"File '{filename}' row {index} protein IDs not found in dictionary, set first protein ID '{first_pid}' as leading protein")
                            continue  # Skip to the next row
                else:
                    if leading_protein not in protein_mapping:
                        local_errors.append(f"File '{filename}' row {index} leading protein '{leading_protein}' not found in dictionary")
                        continue  # Skip to the next row
                    gene_list = protein_mapping[leading_protein][0]
                    gene_name = gene_list[0] if gene_list else None
                    protein_name = protein_mapping[leading_protein][1]
                    df.at[index, 'Gene names'] = gene_name
                    df.at[index, 'Protein names'] = protein_name
            else:
                # Use the first gene name in case of multiple gene names
                gene_name = gene_names.split(';')[0].strip()
                found = False
                for pid, (gene_list, protein) in protein_mapping.items():
                    if gene_name in gene_list:
                        protein_id = pid
                        protein_name = protein
                        df.at[index, 'Leading proteins'] = protein_id
                        df.at[index, 'Protein names'] = protein_name
                        found = True
                        break
                if not found:
                    local_errors.append(f"File '{filename}' row {index} gene name '{gene_name}' not found in dictionary")
                    continue  # Skip to the next row

        # Save the updated DataFrame back to the .pkl file
        df.to_pickle(file_path)

    except Exception as e:
        local_errors.append(f"Error processing file '{filename}': {str(e)}")

    return local_errors

# Get the list of .pkl files
pkl_files = [f for f in os.listdir(directory) if f.endswith('.pkl')]

# Process files in parallel
all_errors = Parallel(n_jobs=15)(delayed(process_file)(filename) for filename in pkl_files)

# Flatten the list of errors
errors = [item for sublist in all_errors for item in sublist]

# Print all errors
if errors:
    print("Errors encountered during processing:")
    for error in errors:
        print(error)
else:
    print("All files processed successfully.")
```

```python
import re

# Path to the error log file
log_file_path = '/home/vmottaqi/decryptm/curves/processed_curves/modified_processed_curves/test/output.log'  # Replace with the path to your log file
output_file_path = '/home/vmottaqi/decryptm/curves/processed_curves/missing_genes.txt'  # Replace with the path to your output file

# Initialize a set to store unique gene names
missing_genes = set()

# Regular expression to extract gene names from the log lines
gene_name_pattern = re.compile(r"gene name '(.+?)' not found in dictionary")

# Read the log file and extract gene names
with open(log_file_path, 'r') as log_file:
    for line in log_file:
        match = gene_name_pattern.search(line)
        if match:
            missing_genes.add(match.group(1))

```

```python
import pandas as pd
import os
import pickle

def update_drug_info(drug_dict, drug_names, smiles=None):
    # Use the list of drug names directly as the key
    key_tuple = tuple(drug_names)  # Convert list to a tuple to use as a dictionary key
    if key_tuple not in drug_dict:
        drug_dict[key_tuple] = smiles

file_dir = '/home/vmottaqi/kg/drug_metadata'
files = [os.path.join(file_dir, f) for f in os.listdir(file_dir) if '_drug_info.csv' in f]

drug_dict = {}

for file_path in files:
    df = pd.read_csv(file_path, encoding='ISO-8859-1')
    obj_name = os.path.basename(file_path).replace('_drug_info.csv', '')
    name_cols = []
    smiles_col = 'smiles'  # This might not exist in all files

    # Define column configurations for each dataset
    if obj_name == "CCLE":
        name_cols = ['Compound..code.or.generic.name.', 'Compound..brand.name.', 'Compound', 'treatmentid', 'drug.name']
    elif obj_name == "Tavor":
        name_cols = ['Compound.Name', 'treatmentid']
    elif obj_name == "UHNBreast":
        name_cols = ['treatmentid']
    elif obj_name == "PDTX":
        name_cols = ['DRUG_NAME', 'BRAND_NAME', 'unique.treatmentid']
    elif obj_name == "GRAY":
        name_cols = ['treatmentid']
    elif obj_name == "gCSI":
        name_cols = ['treatmentid']
    elif obj_name == "BeatAML":
        name_cols = ['drug.name']
    elif obj_name == "CTRPv2":
        name_cols = ['Master_cpd_id', 'cpd_name', 'broad_cpd_id', 'treatmentid']
    elif obj_name == "FIMM":
        name_cols = ['Drug.name', 'treatmentid']
    elif obj_name == "GBM2":
        name_cols = ['GBM.treatmentid', 'treatmentid', 'Compound.name']
    elif obj_name == "GBM3":
        name_cols = ['GBM.treatmentid', 'treatmentid', 'Compound.name']
    elif obj_name == "GDSCv2":
        name_cols = ['DRUG_NAME', 'SYNONYMS', 'treatmentid']
    elif obj_name == "GDSCv1":
        name_cols = ['DRUG_NAME', 'SYNONYMS', 'treatmentid']
    elif obj_name == "NCI60":
        name_cols = ['treatmentid', 'NCI60.treatmentid']
    elif obj_name == "PRISM":
        name_cols = ['PRISM.treatmentid', 'broad_id', 'treatmentid']
    
    else:
        print(f"No configuration for {obj_name}, skipping...")
        continue

    for index, row in df.iterrows():
        # Collect all drug names from the specified columns for the current dataset
        drug_names = [str(row[col]) for col in name_cols if col in df.columns and pd.notna(row[col])]
        # Get SMILES string if available
        smiles = str(row[smiles_col]) if smiles_col in df.columns and pd.notna(row[smiles_col]) else None
        # Update dictionary with drug names and SMILES
        update_drug_info(drug_dict, drug_names, smiles)

# Serialize the drug dictionary
with open('/home/vmottaqi/kg/drug_metadata/drug_dictionary.pkl', 'wb') as f:
    pickle.dump(drug_dict, f)

print("Drug dictionary has been saved.")
```

```python
import pandas as pd
import pickle

# Load the dictionary
with open('/home/vmottaqi/kg/drug_metadata/drug_dictionary.pkl', 'rb') as f:
    drug_dict = pickle.load(f)

# Process each key in the dictionary
new_drug_dict = {}
for key, smiles in drug_dict.items():
    new_key = set()  # Initialize a new set for the updated keys
    for name in key:
        # Convert to lowercase
        name = name.lower()
        new_key.add(name)  # Add the original name
        if '///' in name:
            parts = name.split('///')
            new_key.update(parts)  # Add each part as a separate entry
    
    # Convert the updated set back to a tuple and assign the smiles
    new_key = tuple(sorted(new_key))
    new_drug_dict[new_key] = smiles  # Update the dictionary with the processed keys

# Serialize the updated drug dictionary
with open('/home/vmottaqi/kg/drug_metadata/drug_dictionary_v2.pkl', 'wb') as f:
    pickle.dump(new_drug_dict, f)

print("Updated drug dictionary has been saved.")
```

```python
import pickle

# Load the dictionary
with open('/home/vmottaqi/kg/drug_metadata/drug_dictionary_v2.pkl', 'rb') as f:
    drug_dict = pickle.load(f)

def find_merge_key(key, key_to_group):
    """ Helper function to find if a key should be merged with an existing group based on overlapping drug names. """
    for existing_key, group_key in key_to_group.items():
        if set(key) & set(existing_key):  # If there is an intersection
            return group_key
    return None

# This dictionary will map every key to its "group" key where it should be merged
key_to_group = {}
merged_drug_dict = {}

for key, smiles in drug_dict.items():
    if smiles is None:
        smiles = []
    elif not isinstance(smiles, list):
        smiles = [smiles]

    merge_key = find_merge_key(key, key_to_group)
    if merge_key:
        # Merge current key's items into the found merge key's items
        merged_drug_dict[merge_key].update(smiles)
        # Update the key mapping to reflect that this key also belongs to the merge group
        key_to_group[key] = merge_key
    else:
        # If no merge key is found, this key becomes a new group by itself
        merged_drug_dict[key] = set(smiles)
        key_to_group[key] = key

# Convert sets to lists for serialization and clean up dictionary keys
final_drug_dict = {tuple(sorted(key)): list(smiles) for key, smiles in merged_drug_dict.items()}

# Serialize the merged drug dictionary
with open('/home/vmottaqi/kg/drug_metadata/drug_dictionary_v3.pkl', 'wb') as f:
    pickle.dump(final_drug_dict, f)

print("Merged drug dictionary has been saved.")
```
